```python
import numpy as np
```

```python
v = np.array([2,4,1])
u = np.array([1,3,5])
```

```python
np.dot(v,u)
```

```python
np.linalg.norm(u,1), np.linalg.norm(u,2), np.linalg.norm(u,3), np.linalg.norm(u,4), np.linalg.norm(u,np.Inf)
```

```python
np.linalg.norm(v,1), np.linalg.norm(v,2), np.linalg.norm(v,3), np.linalg.norm(v,4), np.linalg.norm(v,np.Inf)
```

```python
np.add(v,u)
```

```python
np.linalg.norm(u+v)
```
